```python
labels_online = ["test_registration", "test_credential_checking"]
labels_offline = ["test_key_generation", "test_credential_checking", "test_credential_showing", "test_credential_issuance"]

daniel_online_med  = [47.8663,96.3454]
daniel_online_std  = [12.5008,12.9726]
daniel_offline_med = [26.7376,45.3217,46.9227,44.3740]
daniel_offline_std = [1.6397,3.8484,5.2388,4.2665]
daniel_label = "Intel i7-8565U (8) @ 1.80GHz, 16GB memory"

simon_online_med  = [53.4226,96.0179]
simon_online_std  = [20.2798,2.6315]
simon_offline_med = [25.4516,38.8639,39.5684,39.8717]
simon_offline_std = [1.2599,2.0414,2.1804,1.9087]
simon_label = "Intel i9-9880H (16) @ 2.30GHz, 16GB memory"

title = "ABC system benchmark on 500 rounds"

title_online  = "Online, "
title_offline = "Offline, "

units = "ms"
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.gcf().subplots_adjust(bottom=0.15)
```

```python
x_online = list(range(len(daniel_online_med)))
x_offline = list(range(len(x_online),len(x_online)+len(daniel_offline_med)))

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#online
ax.errorbar([x-0.1 for x in x_online], daniel_online_med, daniel_online_std, linestyle='None', marker='o', label=title_online+daniel_label, color="b")
ax.errorbar([x+0.1 for x in x_online], simon_online_med, simon_online_std, linestyle='None', marker='o', label=title_online+simon_label, color="r")

#offline
ax.errorbar([x-0.1 for x in x_offline], daniel_offline_med, daniel_offline_std, linestyle='None', marker='o', label=title_offline+daniel_label, color="g")
ax.errorbar([x+0.1 for x in x_offline], simon_offline_med, simon_offline_std, linestyle='None', marker='o', label=title_offline+simon_label, color="y")

ax.legend()

ax.set_yticks(np.arange(0, 150, 20))
ax.set_yticks(np.arange(0, 150, 5), minor=True)
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

plt.xticks(x_online+x_offline, labels_online+labels_offline, rotation=20, ha='right')
plt.ylabel(units, rotation="horizontal")
plt.title(title)
plt.savefig(title.replace(" ", "_")+".png", dpi=300, bbox_inches='tight')

plt.show()
```
